```python
import pandas as pd
```

```python
usuarios = pd.read_excel('MATRIZ JANEIRO 2020 DIRCEU.xlsx')
bd = pd.read_excel('USUARIOS_DEZEMBRO.xlsx')
```

```python
bd.set_index('matricula', inplace=True)
usuarios.set_index('matricula', inplace=True)
```

```python
cidade = bd['cidade']
```

```python
cidade = cidade.to_dict()
```

```python
usuarios['cidade'] = usuarios.index.map(cidade)
```

```python
usuarios.to_excel("cidade.xlsx")
```
