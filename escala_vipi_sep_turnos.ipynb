```python
import pandas as pd
from datetime import date
import random
```

```python
dias = tuple(range(1, 8))
diasDaSemana = ('Domingo', 'Segunda-feira', 'Terça-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sábado')
mes = []
mesAtual = 4

for dia in dias:
    hoje = date(2020, mesAtual, dia)
    mes.append([dia, diasDaSemana[hoje.weekday()]])

domingos = []
for dia in mes:
    if dia[1] == 'Domingo':
        domingos.append(dia[0])
```

```python
import numpy as np
def setdiff_sorted(array1,array2,assume_unique=False):
    ans = np.setdiff1d(array1,array2,assume_unique).tolist()
    if assume_unique:
        return sorted(ans)
    return ans
```
