```python
import pycep_correios
```

```python
def cidade(cep):
    try:
        endereco = pycep_correios.consultar_cep(cep)
        return endereco['cidade']
    except:
        return 'NaN'
```

```python
cidades = []
for each in aracaju['cep']:
    cep = str(each)
    city = cidade(cep)
    print(city)
    cidades.append(city)
```

```python
aracaju['cidade'] = cidades
```
