```python
dias = tuple(range(1, 8))
diasDaSemana = ('Segunda-feira', 'Terça-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sábado', 'Domingo')
mes = []
mesAtual = 4

for dia in dias:
    hoje = date(2020, mesAtual, dia)
    mes.append([dia, diasDaSemana[hoje.weekday()]])

domingos = []
for dia in mes:
    if dia[1] == 'Domingo':
        domingos.append(dia[0])
```
