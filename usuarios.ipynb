```python
import pandas as pd
import numpy as np
import math
import datetime as dt
now = dt.datetime.now() # current date and time
data = now.strftime("%Y_%m_%d")
```

```python
usuarios = pd.read_excel('bd_março.xlsx')
ativos = pd.read_excel('2020_03_20_usuarios_ativos.xlsx')
```

```python
usuarios
```

```python
usuarios['status_frequencia_atual'] = usuarios['matricula'].isin(ativos['matricula'])
```

```python
usuarios.head()
```

```python
def label(element):
    if element:
        return 'ATIVO'
    else:
        return 'INATIVO'
usuarios['status_frequencia_atual'] = usuarios['status_frequencia_atual'].apply(label)
usuarios['STATUS_FREQUENCIA_ANTERIOR'] = usuarios['status_frequencia']
usuarios['status_frequencia'] = usuarios['status_frequencia_atual']
usuarios = usuarios.drop(['status_frequencia_atual'], axis=1)
usuarios
```

```python
titulo1 = 'bd_março.xlsx'
usuarios.to_excel(titulo1, index=False)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
ativos = 0
inativos = 0
for each in usuarios['status_frequencia']:
    if each == 'ATIVO':
        ativos += 1
    else:
        inativos += 1
# ativos = len(usuarios['status_frequencia'] == 'ATIVO')
# inativos = len(usuarios['status_frequencia'] == 'INATIVO')
total = ativos + inativos
```

```python
# Pie chart
labels = ['Ativos', 'Inativos']
sizes = [ativos, inativos]

explode = (0, 0.1)
# cores
colors = ['#ff9999','#66b3ff']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
plt.tight_layout()

plt.savefig('ativos_inativos.png', bbox_inches='tight')
plt.show()
plt.close()
```
